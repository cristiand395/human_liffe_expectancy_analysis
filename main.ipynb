# Human Life Expectancy
In this notebook I`m going to analyze Human Life Expectancy in the different countries from 1990 to 2019.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

## 1. Importing Data

```python
dataset = pd.read_csv('data/Human_life_Expectancy.csv')
dataset
```

## 2. Cleaning Data

```python
dataset.isnull().sum()
```

```python
dataset.dtypes
```

```python
# Select countries for the entire region, droping their sub regions
dataset = dataset[dataset['Region'] == 'Total']
dataset
```

```python
dataset.iloc[4].dtypes
```

```python
# Becausa I´m going to analyze for the last 20 years, the rows that have `Not Available` data will be dropped
dataset[dataset['1995'] == 'Not Available']
```

```python
for i in range(1990, 2020):
    i = str(i)
    dataset = dataset[dataset[i] != 'Not Available']
dataset
```

```python
# CHange objets data from years columns to float numbers
for i in range(1990, 2020):
    i = str(i)
    dataset[i] = dataset[i].astype(float)
```

```python
dataset.dtypes
```

```python
dataset
```

```python
dataset = dataset.drop(['Level','Region','Country_Code'], axis = 1)
dataset
```

```python
dataset.reset_index(drop = True, inplace = True)
dataset
```

## 3. Plotting Clean Data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```
